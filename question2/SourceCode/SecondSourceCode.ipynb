```python
#Load the required library 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Reading in a CSV file of student performance data into a pandas dataframe called "student_data"
student_data = pd.read_csv("StudentsPerformance.csv")

#Dropping any rows with missing values using the dropna()function
student_data = student_data.dropna()

#Renaming the columns of the dataframe to make them more readable using the rename()function
student_data = student_data.rename(columns={
    "gender": "Gender",
    "race/ethnicity": "Race_Ethnicity",
    "parental level of education": "Parent_Education",
    "lunch": "Lunch",
    "test preparation course": "Test_Preparation",
    "math score": "Math_Score",
    "reading score": "Reading_Score",
    "writing score": "Writing_Score"})

#Removing any leading or trailing white spaces from the columns using the str.strip()function
student_data["Gender"] = student_data["Gender"].str.strip()
student_data["Race_Ethnicity"] = student_data["Race_Ethnicity"].str.strip()
student_data["Parent_Education"] = student_data["Parent_Education"].str.strip()
student_data["Lunch"] = student_data["Lunch"].str.strip()
student_data["Test_Preparation"] = student_data["Test_Preparation"].str.strip()

#Converting the "Race/Ethnicity" column to a categorical variable using the pd.Categorical()function
student_data["Race_Ethnicity"] = pd.Categorical(student_data["Race_Ethnicity"])

#Saving the cleaned dataframe to a CSV file called "clean_data.csv"
student_data.to_csv("clean_data.csv", index=False)

#Plot Graph 1
sns.scatterplot(data=student_data, x="Math_Score", y="Reading_Score", hue="Gender")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.title("Math Score vs Reading Score by Gender")
```

```python
#Plot graph 2
sns.boxplot(data=student_data, x="Race_Ethnicity", y="Math_Score", hue="Race_Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Math Score")
plt.title("Math Score by Race/Ethnicity")
```

```python
#Plot graph 3
sns.countplot(data=student_data, x="Parent_Education", hue="Test_Preparation")
plt.xlabel("Parental Level of Education")
plt.ylabel("Count")
plt.title("Test Preparation Course by Parental Level of Education")
plt.legend(title="Test Preparation Course")
```

```python
#Plot graph 4
sns.kdeplot(data=student_data, x="Writing_Score", hue="Gender")
plt.xlabel("Writing Score")
plt.ylabel("Density")
plt.title("Density of Writing Scores by Gender")
```

```python
#Plot graph 5
sns.countplot(data=student_data, x="Race_Ethnicity", hue="Lunch")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Count")
plt.title("Lunch by Race/Ethnicity")
plt.legend(title="Lunch")
```
